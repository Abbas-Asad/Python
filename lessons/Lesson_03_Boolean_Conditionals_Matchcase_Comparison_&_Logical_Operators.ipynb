# Match case

```python
# 1. Basic Calculator
num_first: float = float(input("Enter first number: "))
num_second: float = float(input("Enter second number: "))
operation: str = input("Select an operation: ( + , - , * , / ) ")

match operation:
  case "+":
    print(num_first + num_second)
  case "-":
    print(num_first - num_second)
  case "*":
    print(num_first * num_second)
  case "/":
    print(num_first / num_second)
  case _: # Default case (like else)
    print("Invalid Operation!")
```

```python
# 2. Grade checker
def grade_student(grade):
  match grade:
    case "A":
      print("Your grade is A")
    case "B":
      print("Your grade is B")
    case "C":
      print("Your grade is C")
    case _: # Default case (like else)
      print("Invalid grade")

grade_student("A") # Output: Your grade is A
# grade_student("D") # Output: Invalid grade
```

```python
# 3. Day checker
day_num = int(input("Enter a number (1-7): "))

match day_num:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("Invalid day number!")
```

## Important Note:  

If we're using match-case in Python, we must match exact values directly.
We cannot use programming logic like list indexing or lookups inside case statements. If we try that way, we lose the benefit of case _: (default/fallback). So now we understand that match-case is best for matching exact values only.
