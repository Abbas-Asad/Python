# Dictionary Comprehensions

```python
fruits_price = {
    "apple": 100,
    "banana": 50,
    "cherry": 90,
    "mango": 150
}

expensive_fruits = {k: v for k, v in fruits_price.items() if v >= 100}
print(expensive_fruits)
```

```python
countries_capitals = {
    "Pakistan": "Islamabad",
    "India": "New Delhi",
    "USA": "Washington D.C.",
    "UK": "London"
}

countries_cap_startswith_i = {k: v for k, v in countries_capitals.items() if v.startswith("I")}
print(countries_cap_startswith_i)
```

```python
cities_temperature_in_celsius = {
    "Karachi": 34,
    "Lahore": 37,
    "Islamabad": 30,
    "Quetta": 28,
    "Peshawar": 35,
    "Multan": 38
}

cities_temperature_in_fahrenheit = {k: ((v * 9/5) + 32) for k, v in cities_temperature_in_celsius.items()}
print(f"The temperature in celsius is: {cities_temperature_in_celsius}")
print(f"The temperature in fahrenheit is: {cities_temperature_in_fahrenheit}")
```

## Break & Continue

```python
# Break example
for num in range(1,11):
  if num == 8:
    break
  print(num)

# Continue example
for num in range(1,101):
  if num % 10 == 0:
    continue
  print(num)

# We use break for breaking loop from somewhere
# We use continue to skip the particular iteration and let loop continue from next iteration
```
