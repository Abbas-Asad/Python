# Generator (Tokens in LLM)

```python
# First cell: define the generator and initialize it
def my_counter():
    i = 0
    while True:
        yield i
        i += 1

# global generator object
gen = my_counter()
```

```python
print(next(gen))
```

```python
def generator():
  yield 1
  yield 2
  yield 3

gen = generator()
```

```python
print(next(gen))
```

```python
def simple_generator():
    yield 1
    yield 2
    yield 3

# Create a generator object
gen = simple_generator()

print(gen, " : ", type(gen))

# Iterate over the generator
for value in gen:
    print(value, " : ", type(value))
```

```python
try:
  print(next(gen)) #error: StopIteration
except BaseException as e:
  print(type(e))
```

# LEGB Precedence Example

```python
# L(local), E(enclosed), G(global), B(Built in)

from math import *

print("Math: pi           = ", pi)

pi = 1
print("Global: pi         = ", pi)

class MyClass:
  pi = 2
  print("MyClass: pi        = ",pi)

  def my_function():
    pi = 3
    print("my_function: pi    = ",pi)

    def inner_function():
      pi = 4
      print("inner_function: pi = ",pi)


    inner_function()

  my_function()
```

# global & nonlocal keywords

```python
# prompt: I want a basic example to demonstrate global and nonlocal keywords

# # global & nonlocal keywords
x = 10  # Global variable

def outer_function():
    x = 20  # Enclosed variable (in outer_function)

    def inner_function():
        nonlocal x  # Modify the enclosed variable
        x = 30
        print("Inner function:", x)  # Output: 30

    inner_function()
    print("Outer function:", x)  # Output: 30

    def another_inner():
      global x # Modify the global variable
      x = 40
      print("Another inner:", x) #Output: 40

    another_inner()
    print("Outer function after another inner:", x) #Output: 30


outer_function()
print("Global scope:", x)  # Output: 40
```

# Overload Decorator

```python
# Overloads are just for type checkers, they ends with ellipsis ... they don't have actual implementation
from typing import overload, Literal, Union

@overload
def get_discount(amount: float, type: Literal["percent"], value: float) -> float: ...
@overload
def get_discount(amount: float, type: Literal["flat"], value: float) -> float: ...

def get_discount(amount: float, type: Union[Literal["percent"], Literal["flat"]], value: float) -> float:
    if type == "percent":
        return amount - (amount * value / 100)
    elif type == "flat":
        return amount - value
    else:
        raise ValueError("Invalid discount type")

print(get_discount(1000, "percent", 10))
print(get_discount(1000, "flat", 200))
```

```python
from typing import overload, Literal, Union

@overload
def get_area(shape: Literal["circle"], radius: float) -> float: ...
@overload
def get_area(shape: Literal["square"], side: float) -> float: ...
@overload
def get_area(shape: Literal["rectangle"], length: float, width: float) -> float: ...
@overload
def get_area(shape: Literal["triangle"], base: float, height: float) -> float: ...

def get_area(shape: str, *args: float) -> float:
    if shape == "circle":
        radius, = args
        return 3.1416 * radius * radius
    elif shape == "square":
        side, = args
        return side * side
    elif shape == "rectangle":
        length, width = args
        return length * width
    elif shape == "triangle":
        base, height = args
        return 0.5 * base * height
    else:
        raise ValueError("Unknown shape")


print("Circle:", get_area("circle", 5))
print("Square:", get_area("square", 4))
print("Rectangle:", get_area("rectangle", 4, 6))
print("Triangle:", get_area("triangle", 4, 5))
```
