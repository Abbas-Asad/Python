```python
import asyncio, time

def make_zinger():
  print("Zinger is preparing")
  time.sleep(3)
  print("Zinger is ready")

def make_limca():
  print("Limca is preparing")
  time.sleep(3)
  print("Limca is ready")

def main():
  make_zinger()
  make_limca()

main()
```

```python
import asyncio, time

async def make_zinger():
  print("Zinger is preparing")
  await asyncio.sleep(3)
  print("Zinger is ready")

async def make_limca():
  print("Limca is preparing")
  await asyncio.sleep(3)
  print("Limca is ready")

async def make_exception():
  print("Limca is preparing")
  # await asyncio.sleep(3)
  # print("Limca is ready")
  raise ValueError("This is async value error")

async def main():
  task_1 = asyncio.create_task(make_zinger())
  task_2 = asyncio.create_task(make_limca())
  task_3 = asyncio.create_task(make_exception())
  print("===========Start===========")
  await asyncio.gather(task_3, task_1, task_2)
  print("===========End===========")

asyncio.run(main())
```

```python
import asyncio

async def task_a():
    await asyncio.sleep(1)
    print("✅ Task A done")

async def task_b():
    await asyncio.sleep(2)
    raise RuntimeError("❌ Task B failed!")

async def task_c():
    await asyncio.sleep(3)
    print("✅ Task C done")

async def main():
    await asyncio.gather(
        task_a(),
        task_b(),  # Yeh fail karega
        task_c()
    )

asyncio.run(main())
```
