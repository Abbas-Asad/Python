# Lists

```python
# List of Fruits
fruits = ["Apple", "Banana", "Cherry", "Date", "Elderberry"]

# check if item exists
print("Apple" in fruits)

# slicing (before last index)
print(fruits[0:4])

# checking length
print(len(fruits))
```

#List Comprehension

```python
students_marks: list = [40, 70, 88, 96, 89]

increased_marks: list = [x + 15 for x in students_marks if x <= 85]
print(increased_marks)
```

```python
nums: list = list(range(10))

squared_nums: list = [x**2 for x in nums]
print(squared_nums)
```

```python
nums: list = list(range(1,21))

even_nums: list = [x for x in nums if x % 2 == 0]
print(even_nums)
```

```python
words: list = ['apple', 'ball', 'cat', 'dog']

first_letter: list = [x for x in words]
print(first_letter)
```
