# Try & Except

```python
# 1
try:
  print(x)

except Exception as err:
  print(f"First Error is: {err}")

# 2
try:
  print(10 / 0)

except Exception as err:
  print(f"Second Error is: {err}")
```

# Custom Exceptions

```python
class TenAboveError(Exception):
  "Custom exception for above 10"
  pass

def check_num(num):
    if num > 10:
        raise TenAboveError("Numbers above 10 are not allowed!")
    else:
        print(f"{n} is < 10")

try:
  check_num(11)
except TenAboveError as e:
  print(e)
```
