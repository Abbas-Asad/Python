Generics ka matlab hai: ek aisi cheez design karna jo multiple types ke liye kaam kare bina har type ke liye alag se likhne ke.

Ye concept typing module ke through Python mein implement hota hai. Ismein hum TypeVar use karte hain.

```python
from typing import TypeVar

T = TypeVar('T')  # Ye ek generic type hai

def identity(item: T) -> T:
    return item

print(identity([3,4,6,7]))
print(identity(["abbas","asad","hamza"]))
print(identity([True, False]))
```

```python
from typing import TypeVar, List

T = TypeVar('T')  # Ye ek generic type hai
def first_item(items: List[T]) -> T:
    return items[0]

first_item(["abbas","saleem","hamza"])
```

# Bound TypeVar

```python
from typing import TypeVar

Number = TypeVar('Number', int, float)

def add(a: Number, b: Number) -> Number:
    return a + b

add(2, 8.5)
```

# Bound Second Way

```python
from typing import TypeVar

T = TypeVar('T', bound=int)  # Sirf int (ya uske subclasses) allowed

def max_of_two(a: T, b: T) -> T:
    return a if a > b else b

# print(max_of_two(5, 10))      # ✅ valid
# print(max_of_two(True, False)) # ✅ valid (bool is subclass of int)
# print(max_of_two("a", "b"))  ❌ invalid — str not allowed
```
