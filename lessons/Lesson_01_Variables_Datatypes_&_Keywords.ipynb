# Variables, Data types & Keywords

```python
name = "abbas"
print("\nname:", name)

age = 22.5
print("\nage:", age)

roll_no = 857896
print("\nroll_no:", roll_no)

studied_from = ["aptech", "domino", "google"]
print("\nstudied_from:", studied_from)
```

## Important Points about data types:
1. All these types will show in type(value) as <class 'typename'>

2. List and tuple can contain any combination of types inside.

3. Set and frozenset automatically remove duplicates.

4. Dict uses key-value pairs, where keys must be unique.

5. Boolean values are a subtype of integers (True == 1, False == 0).

6. Binary types are used mostly for working with files, networking, etc.

```python
# Python Data Types ‚Äì All-in-One Practice Script

print("üî¢ Numeric Types")
my_int = 10
print("int:", type(my_int), my_int)

my_float = 10.5
print("float:", type(my_float), my_float)

my_complex = 3 + 5j
print("complex:", type(my_complex), my_complex)

print("\nüßµ Text Type")
my_str = "Hello, Python!"
print("str:", type(my_str), my_str)

print("\nüß∫ Sequence Types")
my_list = [1, 2, 3, "apple"]
print("list:", type(my_list), my_list)

my_tuple = (1, 2, 3, "banana")
print("tuple:", type(my_tuple), my_tuple)

my_range = range(5)
print("range:", type(my_range), list(my_range))

print("\nüóÇÔ∏è Mapping Type")
my_dict = {"name": "Abbas", "age": 22}
print("dict:", type(my_dict), my_dict)

print("\nüì¶ Set Types")
my_set = {1, 2, 3, 1, 2}
print("set:", type(my_set), my_set)

my_frozenset = frozenset([1, 2, 3])
print("frozenset:", type(my_frozenset), my_frozenset)

print("\n‚úÖ Boolean Type")
my_bool = True
print("bool:", type(my_bool), my_bool)

print("\nüö´ None Type")
my_none = None
print("NoneType:", type(my_none), my_none)
```

```python
def check_type(value):
    print(type(value))

# Test it
check_type("hello")         # <class 'str'>
check_type(123)             # <class 'int'>
check_type([1, 2, 3])       # <class 'list'>
check_type(None)            # <class 'NoneType'>
check_type(b"bytes")        # <class 'bytes'>
```

## Keywords

Keywords in Python are `reserved words` that have special meanings and serve specific purposes in the language syntax. Python keywords cannot be used as the names of variables, functions, and classes or any other identifier.

```python
import keyword

# Line continuation (`\`) allows printing a statement over multiple lines, improving code readability without breaking the string.
print("The list of \
keywords is : ")

# printing all keywords at once using "kwlist()"
print(keyword.kwlist)
```
